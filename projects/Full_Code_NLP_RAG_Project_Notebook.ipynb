```python
# Load the model
llm = Llama(
    model_path=model_path,
    n_ctx=5000,
    n_gpu_layers=38,
    n_batch=512
)
```

```python
# function to generate llm response
def response(query,max_tokens=128,temperature=0,top_p=0.95,top_k=50):
    model_output = llm(
      prompt=query,
      max_tokens=max_tokens,
      temperature=temperature,
      top_p=top_p,
      top_k=top_k
    )

    return model_output['choices'][0]['text']
```

### Retriever

```python
# initialize retriever
retriever = vectorstore.as_retriever(
    search_type='similarity',
    search_kwargs={'k': 2}
)
```

```python
# fetch relevant docs with retriever
rel_docs = retriever.get_relevant_documents("How to diagnose appendicitis?")
for doc in rel_docs:
    print(doc.page_content + "\n")
```

- Content relevant to the topic (Appendicitis) is retrieved.

### System and User Prompt Template

```python
# define system prompt
qna_system_message = """
  You are an expert medical information assistant.
  Your primary goal is to provide clear, accurate, and concise answers to medical questions based on the provided context.
  You must answer questions strictly based on the context provided, adhering to medical standards.
  If the answer is not in the context, clearly state that you don't have information on that topic.
  Do not hallucinate or provide your own medical advice.
  Always summarize your answers in a concise and clear manner.
"""
```

```python
# user input prompt template
qna_user_message_template = """
  **Medical Context:** {context}

  **Question:** {question}
"""
```

### Response Function

```python
def generate_rag_response(user_input,k=3,max_tokens=128,temperature=0,top_p=0.95,top_k=50):
    global qna_system_message,qna_user_message_template
    # Retrieve relevant document chunks
    relevant_document_chunks = retriever.get_relevant_documents(query=user_input,k=k)
    context_list = [d.page_content for d in relevant_document_chunks]

    # Combine document chunks into a single context
    context_for_query = ". ".join(context_list)

    user_message = qna_user_message_template.replace('{context}', context_for_query)
    user_message = user_message.replace('{question}', user_input)

    prompt = qna_system_message + '\n' + user_message

    # Generate the response
    try:
        response = llm(
                  prompt=prompt,
                  max_tokens=max_tokens,
                  temperature=temperature,
                  top_p=top_p,
                  top_k=top_k
                  )

        # Extract and print the model's response
        response = response['choices'][0]['text'].strip()
    except Exception as e:
        response = f'Sorry, I encountered the following error: \n {e}'

    return response
```

### **Observations**

1.  **max_tokens (Maximum Response Length):**
    * Increasing **max_tokens** (e.g., from default to 512) directly addressed the issue of truncated responses observed in initial RAG and LLM-only outputs. It allowed the model to provide more comprehensive answers without being cut off prematurely.
    * When **max_tokens** was insufficient, answers were abruptly terminated, even if the underlying context contained more relevant information. Increasing it ensured that the model could articulate all necessary details derived from the retrieved context.

2.  **k (Number of Retrieved Documents):**
    * Increasing the **k** value (e.g., from default 3 to 5) meant providing the LLM with a larger pool of relevant context chunks from the vector database.
    * A higher **k** value generally led to more detailed and potentially richer responses, as the LLM had access to a broader scope of information. This helped ensure that all nuances of a complex query could be addressed, even if some critical information was spread across multiple chunks. For instance, in sepsis management, increasing **k** brought in more detailed treatment options and monitoring parameters.

3.  **temperature (Creativity/Randomness of Response):**
    * Adjusting **temperature** influenced the determinism and creativity of the generated text.
    * A **temperature=0** (or close to 0, as in the default for initial RAG responses) resulted in very factual, direct, and conservative answers, highly predictable and less prone to varied phrasing. Increasing **temperature** (e.g., to 0.3) introduced more descriptive language and varied sentence structures. While maintaining factual accuracy due to RAG, it allowed for more empathetic or slightly more elaborate explanations, as seen in the fracture query.

4.  **top_p (Nucleus Sampling):**
    * **top_p** controls the diversity of the generated text by sampling from the smallest set of most probable tokens whose cumulative probability exceeds **top_p**.
    * Reducing **top_p** (e.g., from 0.95 to 0.9) slightly increased the vocabulary size and introduced more varied word choices, without significantly altering the core factual content. It allowed the model to explore a wider range of tokens that were still highly probable, contributing to a more natural-sounding response.

5.  **top_k (Top-K Sampling):**
    * **top_k** limits the model's choices to the **k** most likely next tokens.
    * Reducing **top_k** (e.g., from 50 to 25) effectively decreased the vocabulary size and made the model's choices more focused on the most probable words. This was particularly useful when responses were getting cut off or becoming too verbose.

### Suitable Combination of Parameters for Evaluation

Considering the goal of providing clear, accurate, and concise medical information while adhering strictly to context and avoiding speculation, the following parameter combination appears most suitable for evaluation:

*   **max_tokens = 512**: This value consistently allowed for comprehensive answers without truncation, providing sufficient detail for complex medical queries.

*   **k = 5**: Retrieving 5 relevant document chunks provides a robust context for the LLM, minimizing the chance of missing critical information while not overwhelming the model with irrelevant data.

*   **temperature = 0.3**: A slightly higher temperature than 0 (which can be too rigid) introduces a small degree of natural language variation and empathy, making the responses more readable and engaging, without risking factual deviations or hallucinations. It maintains a good balance between creativity and factual adherence.

* **top_p = 0.95** and **top_k = 50**: Reducing these values don't have noticeable impact. Therefore, we can keep them at initial defaults.

- We are using the same Mistral model for evaluation, so basically here the llm is rating itself on how well he has performed in the task.

```python
groundedness_rater_system_message  = """
You are tasked with rating AI generated answers to questions posted by users.
You will be presented a question, context used by the AI system to generate the answer and an AI generated answer to the question.

Evaluation criteria:
The task is to judge the extent to which the metric is followed by the answer.
1 - The metric is not followed at all
2 - The metric is followed only to a limited extent
3 - The metric is followed to a good extent
4 - The metric is followed mostly
5 - The metric is followed completely

Metric:
The answer should be derived only from the information presented in the context

Instructions:
1. First write down the steps that are needed to evaluate the answer as per the metric.
2. Give a step-by-step explanation if the answer adheres to the metric considering the question and context as the input.
3. Next, evaluate the extent to which the metric is followed.
4. Use the previous information to rate the answer using the evaluaton criteria and assign a score.
"""
```

```python
relevance_rater_system_message = """
You are tasked with rating AI generated answers to questions posed by users.
You will be presented a question, context used by the AI system to generate the answer and an AI generated answer to the question.

Evaluation criteria:
The task is to judge the extent to which the metric is followed by the answer.
1 - The metric is not followed at all
2 - The metric is followed only to a limited extent
3 - The metric is followed to a good extent
4 - The metric is followed mostly
5 - The metric is followed completely

Metric:
Relevance measures how well the answer addresses the main aspects of the question, based on the context.
Consider whether all and only the important aspects are contained in the answer when evaluating relevance.

Instructions:
1. First write down the steps that are needed to evaluate the context as per the metric.
2. Give a step-by-step explanation if the context adheres to the metric considering the question as the input.
3. Next, evaluate the extent to which the metric is followed.
4. Use the previous information to rate the context using the evaluaton criteria and assign a score.
"""
```

```python
user_message_template = """
###Question
{question}

###Context
{context}

###Answer
{answer}
"""
```

```python
def generate_ground_relevance_response(user_input,k=3,max_tokens=128,temperature=0,top_p=0.95,top_k=50):
    global qna_system_message,qna_user_message_template
    # Retrieve relevant document chunks
    relevant_document_chunks = retriever.get_relevant_documents(query=user_input,k=3)
    context_list = [d.page_content for d in relevant_document_chunks]
    context_for_query = ". ".join(context_list)

    # Combine user_prompt and system_message to create the prompt
    prompt = f"""[INST]{qna_system_message}\n
                {'user'}: {qna_user_message_template.format(context=context_for_query, question=user_input)}
                [/INST]"""

    response = llm(
            prompt=prompt,
            max_tokens=max_tokens,
            temperature=temperature,
            top_p=top_p,
            top_k=top_k,
            stop=['INST'],
            )

    answer =  response["choices"][0]["text"]

    # Combine user_prompt and system_message to create the prompt
    groundedness_prompt = f"""[INST]{groundedness_rater_system_message}\n
                {'user'}: {user_message_template.format(context=context_for_query, question=user_input, answer=answer)}
                [/INST]"""

    # Combine user_prompt and system_message to create the prompt
    relevance_prompt = f"""[INST]{relevance_rater_system_message}\n
                {'user'}: {user_message_template.format(context=context_for_query, question=user_input, answer=answer)}
                [/INST]"""

    response_1 = llm(
            prompt=groundedness_prompt,
            max_tokens=max_tokens,
            temperature=temperature,
            top_p=top_p,
            top_k=top_k,
            stop=['INST'],
            )

    response_2 = llm(
            prompt=relevance_prompt,
            max_tokens=max_tokens,
            temperature=temperature,
            top_p=top_p,
            top_k=top_k,
            stop=['INST'],
            )

    return response_1['choices'][0]['text'],response_2['choices'][0]['text']
```

    * Continuously monitor and fine-tune the **'k'** parameter. Implement user feedback loop to determine the optimal **'k'** value for different query types, balancing completeness with computational cost.

3.  **Strategic 'max_tokens' Management:**
    * Adequate **max_tokens** are crucial to prevent abrupt termination of responses, ensuring that comprehensive medical information is fully conveyed. A **'max_tokens=512'** was found to be a good balance for this project's queries.
    * Set **max_tokens** generously to allow for detailed explanations, especially for critical medical procedures or conditions. However, implement a summarization layer or allow users to request more detail if initial responses are too brief, rather than defaulting to excessively long outputs.

4.  **Balanced 'temperature' for Readability and Accuracy:**
    * A slightly elevated **'temperature'** (e.g., 0.3) introduces natural language variation without sacrificing factual integrity, making the responses more engaging and readable for healthcare professionals.
    * Maintain a low to moderate **'temperature'** for medical applications. While creativity is generally undesirable in factual medical content, a minimal **'temperature'** can improve readability and user experience by varying sentence structure and phrasing, enhancing comprehension.

5.  **Refined 'top_p' and 'top_k' for Focused Outputs:**
    * Adjusting **'top_p'** and **'top_k'** parameters can help control the vocabulary and focus of the LLM's generation, especially when aiming for conciseness or to avoid irrelevant tangents.
    * Use **'top_k'** and **'top_p'** to fine-tune the model's lexicon, ensuring it remains within relevant medical terminology. This helps prevent the model from generating overly verbose or tangential information, keeping responses concise and on-topic.

6.  **Continuous Evaluation and Feedback Loops:**
    * The **LLM-as-a-judge** evaluation highlighted few areas where the model might still produce content not strictly from the context or miss certain relevant aspects.
    * Implement ongoing evaluation using metrics like groundedness and relevance. Incorporate feedback from medical professionals to continuously refine retrieval mechanisms, prompt engineering, and LLM parameters to improve the quality and clinical utility of responses.
